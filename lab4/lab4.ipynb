```python
import matplotlib.pyplot as plt
import pandas as pd
import warnings
import seaborn as sns
from matplotlib.ticker import ScalarFormatter, AutoMinorLocator

warnings.filterwarnings("ignore", message="Workbook contains no default style")
```

```python
file_path = r'C:\Users\vladi\.vscode\Git\SM_25\SM_25\file1.xlsx'
df1 = pd.read_excel(file_path, engine="openpyxl")

print(df1.head())
```

```python
file_path = r'C:\Users\vladi\.vscode\Git\SM_25\SM_25\file2.xlsx'
df2 = pd.read_excel(file_path, engine="openpyxl")

print(df2.head())
```

```python
df = df1

plt.figure(figsize=(10, 6))
plt.plot(df['nSystem'], df['time'], 'o-', linewidth=2, markersize=8)

x_min = df['nSystem'].min()
x_max = df['nSystem'].max()
x_range = x_max - x_min
plt.xlim(x_min - 0.05 * x_range, x_max + 0.05 * x_range)

plt.xticks(df['nSystem'])

plt.xlabel('Складність моделі (кількість подій)')
plt.ylabel('Час виконання (секунди)')
plt.title('Залежність часу виконання від кількості подій')
plt.grid(True, alpha=0.3)
plt.show()
```
